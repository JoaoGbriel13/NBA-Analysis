```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import lxml
import matplotlib.pyplot as plt
import numpy as np
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.nba.com/stats')
```

```python
button = driver.find_element(By.XPATH, '//*[@id="__next"]/div[2]/div[2]/div[3]/div/div[1]/section[1]/div/div[1]/div/a')
button.click()
```

```python
select_element = driver.find_element(By.XPATH, '//*[@id="__next"]/div[2]/div[2]/div[3]/section[2]/div/div[2]/div[2]/div[1]/div[3]/div/label/div/select')
select = Select(select_element)
select.select_by_index(0)
```

```python
src = driver.page_source
parser = BeautifulSoup(src, "lxml")
```

```python
table = parser.find("table", attrs={'Crom_table__p1iZz'})
```

```python
headers = table.find_all("th")
headersList = [head.text.strip() for head in headers[1:]]
headersCorrected = [head for head in headersList if not "RANK" in head]
print(headersCorrected)
```

```python
rows = table.find_all("tr")[1:]
rowsList = [[td.getText().strip() for td in rows[i].find_all("td")[1:]] for i in range(len(rows))]
print(rowsList)
```

```python
df = pd.DataFrame(data=rowsList, columns=headersCorrected)
```

```python
select_season = driver.find_element(By.XPATH, '//*[@id="__next"]/div[2]/div[2]/div[3]/section[1]/div/div/div[1]/label/div/select')
current_season = Select(select_season)
season_value = current_season.first_selected_option.text
```

```python
df['season'] = season_value
df.head()
```

```python
df = df.infer_objects()
```

```python
df = df.astype(int, errors='ignore')
```

```python
df.head()
```

```python
df['PPG'] = round(df['PTS'] / df['GP'], 2)
df.head()
```

```python
df['2PM'] = df['FGM'] - df['3PM']
df['2PA'] = df['FGA'] - df['3PA']
df.head()
```

```python
arr = np.array([df['2PM'].sum(), df['3PM'].sum()])
labels = ['2PM', '3PM']
plt.title("NBA Totals of 2PM x 3PM in the 22-23 Season")
pie = plt.pie(arr, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
arr = np.array([df['2PA'].sum(), df['3PA'].sum()])
labels = ['2PA', '3PA']
plt.title("NBA Totals of 2PA x 3PA in the 22-23 Season")
pie = plt.pie(arr, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
df_above30 = df.loc[df['Age'] >= 30]
df_above30.head()
```

```python
df_below30 = df.loc[df['Age'] < 30]
df_below30.head()
```

```python
arr = np.array([df_above30['3PM'].sum(), df_below30['3PM'].sum()])
labels = ['Over 30', 'Below 30']
plt.title('NBA 3PM from players over 30 years old vs players under 30 years old')
pie = plt.pie(arr, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
arr = np.array([df_above30['3PA'].sum(), df_below30['3PA'].sum()])
labels = ['Over 30', 'Below 30']
plt.title('NBA 3PA from players over 30 years old vs players under 30 years old')
pie = plt.pie(arr, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
arr = np.array([df_above30['2PA'].sum(), df_below30['2PA'].sum()])
labels = ['Over 30', 'Below 30']
plt.title('NBA 2PA from players over 30 years old vs players under 30 years old')
pie = plt.pie(arr, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
arr = np.array([len(df_above30), len(df_below30)])
labels = ['Over 30', 'Below 30']
plt.title('Percentage of NBA players above 30 years in the NBA 22-23 Season')
pie = plt.pie(arr, labels=labels, autopct='%1.1f%%')
plt.show()
```
